```python
import pyforest
import matplotlib.pyplot as plt
from datetime import datetime
from meteostat import Point, Daily, Stations
```

Source: https://meteostat.net/ & https://github.com/meteostat/meteostat-python . </br>
Used for personal use/educational use.

```python
# Start and end time for graphs
start = datetime(2005, 1, 1)
end = datetime(2005, 12, 31)
```

```python
# Create Point for Amsterdam, NL (Lat, Lon, alt)
location = Point(52.366667, 4.9, -2)
```

```python
# Get daily data for 2005
data = Daily(location, start, end)
data = data.fetch()

# Plot line chart including average, minimum and maximum temperature
data.plot(y=['tavg', 'tmin', 'tmax'])
plt.title('Weather in Amsterdam')
plt.xlabel('Time')
plt.ylabel('Temperature in Celsius')
plt.show()
```
